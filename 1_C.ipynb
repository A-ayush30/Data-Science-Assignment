```python
#importing required libraries
import math
from mpl_toolkits import mplot3d
import pandas as pd
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from google.colab import drive
#importing csv file 
drive.mount('/content/drive')
path = "/content/drive/MyDrive/fods_assign/fods_1.csv"
df = pd.read_csv(path)
#df.describe
cnt = len(df)                                                 #data_count=cnt
```

# Error Contour

```python
w1 = np.linspace(-10, 10, 100)
w2 = np.linspace(-10, 10, 100)
W1,W2 = np.meshgrid(w1, w2)
# fig, ax = plt.subplots(1, 1)
  
Z = 0
for x1, x2, t in df.values:
    Z += np.square(W1*x1+W2*x2 - t)  
Z/=2   

levels = np.linspace(500, 2000,20)              #contour lines
contour = plt.contour(W1, W2, Z, levels=levels)       #levels

#plotting 
plt.clabel(contour, contour.levels, inline=True, fontsize=10)
plt.title('Error Contour Plot')
plt.xlabel('W1')
plt.ylabel('W2')
plt.colorbar(contour)
plt.xlim(-1,2)
plt.ylim(-0.5,3)
plt.draw()
```

# Constraint Regions

```python
#for q=0.5 and eta=1.4
# np.sqrt(abs(w1)) + np.sqrt(abs(w2)) <= eta

eta_n1 = 1.4
w1_linspace_n1 = np.linspace(-2,2,500)

w2_graph_top_n1 = np.square(eta_n1 - np.sqrt(abs(w1_linspace_n1)))
w2_graph_bottom_n1 = -np.square(eta_n1 - np.sqrt(abs(w1_linspace_n1)))

plt.figure()
plt.plot(w1_linspace_n1, w2_graph_top_n1, color = "blue")
plt.plot(w1_linspace_n1, w2_graph_bottom_n1, color = "red")

w2_graph_min_n1 = np.minimum(w2_graph_top_n1, w2_graph_bottom_n1)
w2_graph_max_n1 = np.maximum(w2_graph_top_n1, w2_graph_bottom_n1)

plt.fill_between(w1_linspace_n1, w2_graph_min_n1, w2_graph_max_n1, color="yellow")

plt.xlim((-2.5, 2.5))
plt.ylim((-2, 2))
plt.xlabel('W1')
plt.ylabel('W2')
plt.title("plot: q = 0.5 and $\eta$ = 1.4")
ax = plt.gca()
ax.set_aspect("equal", adjustable = "box")
plt.draw()
```

```python
accumulator = list()
contour = plt.contour(W1, W2, Z, levels=levels)

plt.clabel(contour, contour.levels, inline=True, fontsize=10)

plt.plot(w1_linspace_n1, w2_graph_top_n1, color = "blue")
plt.plot(w1_linspace_n1, w2_graph_bottom_n1, color = "blue")

def f(x):
    Z_1 = 0
    accumulator.append(x)
    for x1,x2,t in df.values:
        Z_1+=np.square((x[0]*x1 + x[1]*x2 - t))
    Z_1/=2
    return Z_1

def constraint(x):
    return np.atleast_1d(1.4 - np.power(abs(x[0]), 0.5) - np.power(abs(x[1]), 0.5))

res_1 = optimize.minimize(f, np.array([0, 1.96]), method="SLSQP", constraints={"fun": constraint, "type": "eq"})

accumulated = np.array(accumulator)

plt.title("Intersection of Tangential Contour and n1")
plt.xlim(-2.2,2.2)
plt.ylim(-2,3)

plt.draw()
```
