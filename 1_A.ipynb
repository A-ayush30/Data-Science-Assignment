```python
# Initially it is given that s is the probability of a customer liking a new update and it is assumed that 
# it follows beta distribution with parameters a,b = (2,2). The posterior distribution of s can be
# obtained by multiplying prior beta distribution (a, b) and likelihood function (l,m) and normalizing it. 
# We can observe that posterior distribution of s will be beta distribution with parameters (a +l,b +m). 
# Where l,m denote the number of people who liked and disliked the update. In total we did three surveys so 
# the beta parameters for the prior and posterior parameters after each survey are 
# After 1st survey: beta(s, a      ,b)       and beta (s,a+40      ,b+10)
# After 2nd survey: beta(s, a+40   ,b+10)    and beta (s,a+40+13   ,b+10+17)
# After 3rd survey: beta(s, a+40+13,b+10+17) and beta (s,a+40+13+70,b+10+17+30)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
# Set the shape parameters
a, b = 2, 2
# Generate the value between
x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
# Plotting the beta distribution after 1st survey
plt.figure(figsize=(7, 7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'g-')
plt.plot(x, beta.pdf(x, a+40, b+10), 'r-')
plt.title('Beta Distribution after 1st survey', fontsize='15')
plt.xlabel('Values of Random Variable S (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python

# Plotting the beta distribution after second survey
plt.figure(figsize=(7, 7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a+40, b+10), 'r-')
plt.plot(x, beta.pdf(x, a+40+13, b+10+17), 'b-')
plt.title('Beta Distribution after 2nd survey', fontsize='15')
plt.xlabel('Values of Random Variable S (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python
# Plotting the beta distribution after third survey
plt.figure(figsize=(7, 7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a+40+13, b+10+17), 'b-')
plt.plot(x, beta.pdf(x, a+40+13+70, b+10+17+30), 'orange')
plt.title('Beta Distribution after 3rd survey', fontsize='15')
plt.xlabel('Values of Random Variable S (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python
#Plotting distribution of s before and after survey.
plt.figure(figsize=(7, 7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'g-')
plt.plot(x, beta.pdf(x, a+40+13+70, b+10+17+30), 'orange')
plt.title('Beta Distribution before and after survey', fontsize='15')
plt.xlabel('Values of Random Variable S (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```
